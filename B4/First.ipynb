```python
import  pandas as pd 
df = pd.read_csv('tips.csv')
df['total_bill'].max()
```

```python
%matplotlib inline
df.plot()
```

```python
df['total_bill'].plot(kind = 'hist')
df['total_bill'].plot(kind = 'hist', grid = True, title = 'Total invoice')
```

```python
df['day'].value_counts().plot(kind = 'bar', 
                              colormap = 'coolwarm', 
                              title = 'The number of visitors by day')
```

```python
df.groupby(by = ['sex', 'smoker'])['tip'].mean().plot(kind = 'bar', grid= True)
```

```python
df.groupby(by = 'day')['total_bill'].agg(['min','max']).plot()
```

```python
df[['total_bill', 'tip']].plot(kind = 'hist', 
                               grid = True, 
                               subplots = True, 
                               title = ['Total invoice', 'Tip amount'], 
                               legend = False)
```

```python
df.plot(x = 'total_bill', 
        y = 'tip', 
        kind = 'scatter', 
        grid = True, 
        title = 'Total invoice VS Tip amount')
```

```python
df.pivot_table(values = ['total_bill', 'tip'], 
               index = 'day', 
               aggfunc = 'mean').plot(kind = 'bar')
```

```python
df.pivot_table(values = 'tip', index = ['sex', 'smoker'], aggfunc = 'count').plot(kind = 'bar')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.hist(df['total_bill'], bins = 20, color = 'green')
axes.set_title('Total invoice')
axes.set_ylabel('The number of orders')
axes.set_xlabel('Order price, USD')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,2])
axes.scatter(x = df['total_bill'], y = df['tip'], marker = 'D')
axes.set_title('Total invoice VS Tip amount')
axes.set_xlabel('Order price')
axes.set_ylabel('Tip amount')
```

```python
tips_per_day = df.groupby(by = 'day')['total_bill'].agg(['min','max']).reset_index()
fig = plt.figure()
main_axes = fig.add_axes([0, 0, 1, 1])
main_axes.hist(df['total_bill'])
insert_axes = fig.add_axes([0.6, 0.6, 0.38, 0.38])
insert_axes.bar(x = tips_per_day['day'], height = tips_per_day['max'])
insert_axes.bar(x = tips_per_day['day'], height = tips_per_day['min'])
```

```python
fig = plt.figure()
main_axes = fig.add_axes([0, 0, 1, 1])
main_axes.hist(df['total_bill'])
insert_axes = fig.add_axes([0.6, 0.6, 0.39, 0.39])
insert_axes.bar(x = tips_per_day['day'], height = tips_per_day['max'], 
                width = 0.4, align = 'edge')
insert_axes.bar(x = tips_per_day['day'], height = tips_per_day['min'], 
                width = -0.4, align = 'edge')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim(0, 60)
axes.bar(x = tips_per_day['day'], height = tips_per_day['max'], width = 0.4, align = 'edge', label = "Maximum Check Amount")
axes.bar(x = tips_per_day['day'], height = tips_per_day['min'], width = -0.4, align = 'edge', label = "Minimum Check Amount")
axes.legend(loc = 1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4))
axes[0].hist(df['total_bill'])
axes[0].set_title('Total invoice')

axes[1].hist(df['tip'])
axes[1].set_title('Tip amount')

axes[2].scatter(x = df['total_bill'], y = df['tip'], marker = 'o')
axes[2].set_title('Total invoice VS Tip amount')
```

```python
import numpy as np 
x = np.linspace(start=-3., stop=3., num=1000)
y1 = np.exp(x)
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y1) 
y2 = 10 + 0.4*x - 0.3*x**2 + 0.1*x**3
axes.plot(x[500:], y1[500:]) 
axes.plot(x, y2)
```

```python
fig = plt.figure(figsize=(7, 4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y1, linestyle='--')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x[::30], y2[::30], 'mD')
```
